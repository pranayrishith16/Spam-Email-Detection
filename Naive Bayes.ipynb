```python
import glob
import os
from nltk.corpus import names
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
from collections import defaultdict

# import nltk
# nltk.download('names')
# nltk.download('wordnet')
# nltk.download('omw-1.4')
```

```python
def letters_only(astr):
    return astr.isalpha()
```

```python
emails,labels = [],[]
file_path = '/kaggle/input/enron-spam/enron4/spam/'
for filename in glob.glob(os.path.join(file_path,'*.txt')):
    with open(filename, 'r', encoding="ISO-8859-1") as infile:
        emails.append(infile.read())
        labels.append(1)
```

```python
file_path = '/kaggle/input/enron-spam/enron4/ham/'
for filename in glob.glob(os.path.join(file_path,'*.txt')):
    with open(filename, 'r', encoding="ISO-8859-1") as infile:
        emails.append(infile.read())
        labels.append(0)
```

```python
def clean_text(docs):
    cleaned_docs = []
    for doc in docs:
        cleaned_docs.append(" ".join([lemmatizer.lemmatize(word.lower()) for word in doc.split() if letters_only(word) and word not in all_names]))
    return cleaned_docs
```

```python
all_names = set(names.words())
lemmatizer = WordNetLemmatizer()

cleaned_emails = clean_text(emails)
print(cleaned_emails[0])
```

```python
cv = CountVectorizer(stop_words="english",max_features=500)
term_docs = cv.fit_transform(cleaned_emails)

feature_names = cv.get_feature_names_out()
print(feature_names)
```

```python
def get_label_index(labels):
    label_index = defaultdict(list)
    for index,label in enumerate(labels):
        label_index[label].append(index)
    return label_index

label_index = get_label_index(labels)
```
